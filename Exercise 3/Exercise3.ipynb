Use logistic regression to predict the probability of a passenger surviving.

```python
import pandas as pd
```

## Dataset File

```python
dataset_url = 'https://github.com/robitussin/CCMACLRL_EXERCISES_OLD/blob/f5fdfba1aabbbafcf2fe49ca6e1c8f92f69d1458/datasets/titanic/train.csv?raw=true'
df = pd.read_csv(dataset_url)
df.head()
```

```python
df.describe()
```

## Test File

```python
test_url = 'https://github.com/robitussin/CCMACLRL_EXERCISES_OLD/blob/f5fdfba1aabbbafcf2fe49ca6e1c8f92f69d1458/datasets/titanic/test.csv?raw=true'
dt=pd.read_csv(test_url)
```

```python
dt.info()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
dt['Age'].fillna(dt['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop('Cabin', axis=1, inplace=True)
dt.drop('Cabin', axis=1, inplace=True)
dt['Fare'].fillna(dt['Fare'].mean(), inplace=True)
```

```python
df['FamilySize'] = df['SibSp'] + df ['Parch']+1
dt['FamilySize'] = dt['SibSp'] + dt ['Parch']+1

df['IsAlone'] = 0
df.loc[df['FamilySize'] ==1, 'IsAlone'] = 1
dt['IsAlone'] = 0
dt.loc[dt['FamilySize'] ==1, 'IsAlone'] = 1

display(df.head())
display(dt.head())
```

```python
df = pd.get_dummies(df, columns=['Sex', 'Embarked', ], prefix=['Sex', 'Embarked'])
dt = pd.get_dummies(dt, columns=['Sex', 'Embarked', ], prefix=['sex', 'Embarked'])
display(df.head())
display(dt.head())
```

```python
features = ['Pclass', 'Age', 'Fare', 'FamilySize', 'IsAlone', 'Sex_female', 'Sex_male', 'Embarked_C', 'Embarked_Q', 'Embarked_S']
```

```python
features = ['Pclass', 'Age', 'Fare', 'FamilySize', 'IsAlone', 'Embarked_C', 'Embarked_Q', 'Embarked_S']

x = df[features]
y = df['Survived']
x_test = dt[features]
display(x.head())
display(y.head())
display(x_test.head())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model=LogisticRegression()
model.fit(x,y)
y_pred=model.predict(x)
accuracy=accuracy_score(y,y_pred)
print("Accuracy:", accuracy)
```

```python
missing_cols = set(x.columns) - set(x_test.columns)
for c in missing_cols:
    x_test[c] = 0
x_test = x_test[x.columns]

id = dt['PassengerId'] # Get PassengerId from the test set
y_pred = model.predict(x_test) # Predict on the test set

# Create a submission DataFrame
submission_df = pd.DataFrame({
    'PassengerId': id, # Correct column name to match the sample submission
    'Survived': y_pred
})

# Save the submission DataFrame to a CSV file
submission_df.to_csv('submission_file.csv', index=False)
print("Submission file created: submission_file.csv")
```
